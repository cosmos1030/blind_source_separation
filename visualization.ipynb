```python
import numpy as np
from scipy.io import wavfile
from scipy.signal import stft
import matplotlib.pyplot as plt

fs, x = wavfile.read('MiniLibriMix/train/mix_clean/14-212-0013_7926-254949-0054.wav')
x = x.astype(float)

f, t, Zxx = stft(x, fs, nperseg=1024, noverlap=512)
V = np.abs(Zxx)

plt.figure()
plt.pcolormesh(t, f, V)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.title('Magnitude Spectrogram of Mixed Signal')
plt.tight_layout()
plt.show()
```
